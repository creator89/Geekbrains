```python
import numpy as np
import pandas as pd
```

```python
authors = pd.DataFrame({'Authors_id':[1, 2, 3], 'Authors_name':['Тургенев', 'Чехов', 'Островский']}, columns = ['Authors_id', 'Authors_name'])
```

```python
book = pd.DataFrame({'Authors_id':[1, 1, 1, 2, 2, 3, 3], 'Book_title':['Отцы и дати', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 'Price':[450, 300, 350, 500, 450, 370, 290]}, columns = ['Authors_id', 'Book_title', 'Price'])
```

```python
#Задание №2 - Объединение Датафреймов по столбцу 'Authors_id'
authors_price = pd.merge(authors, book, on = 'Authors_id', how = 'inner')
```

```python
authors_price
```

```python
#Задние №3 - 5 самых дорогих книг
top5 = (authors_price.sort_values(by = 'Price')).tail()
top5
```

```python
#или так
top5_2 = authors_price.nlargest(5, 'Price')
top5_2
```

```python
#Задание №4 - не корректно работает, по отдельности разобрался с функцией, а ума что то не дал в итоге
authors_stat = pd.DataFrame({'Authors_name': authors_price['Authors_name'], 'min_price': authors_price['Price'].min(), 'max_price': authors_price['Price'].min(), 'mean_price': authors_price['Price'].median()}, columns = ['Authors_name', 'min_price', 'max_price', 'mean_price'])
authors_stat
```

```python
#Задание №5
cover_list = ['Твёрдая', 'Мягкая', 'Мягкая', 'Твёрдая', 'Твёрдая', 'Мягкая', 'Мягкая']
authors_price['cover'] = cover_list
book_info = pd.pivot_table(authors_price, values='Price', index=['Authors_name', 'Book_title'], columns=['cover'], aggfunc=np.sum)
book_info
```

```python
#Замещение пустых значений нулями
book_info['Мягкая'].fillna(0, inplace = True)
book_info['Твёрдая'].fillna(0, inplace = True)
book_info
```

```python
#Импорт и экспорт в пикл
book_info.to_pickle('book_info.pkl')
book_info2 = pd.read_pickle('book_info.pkl')
book_info2
```
